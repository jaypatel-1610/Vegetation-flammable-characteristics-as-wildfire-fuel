```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Final_Vegetation_Dataset(3).csv')

# Set a global green theme for Seaborn
sns.set_palette(sns.color_palette("Greens"))

# Univariate Analysis
# Numerical Variables
numerical_columns = data.select_dtypes(include=['float64']).columns

# Histograms for Numerical Variables
fig, axes = plt.subplots(len(numerical_columns) // 3 + 1, 3, figsize=(15, 15))
axes = axes.flatten()

# Iterate over numerical columns and plot
for i, col in enumerate(numerical_columns):
    sns.histplot(data[col], kde=True, ax=axes[i], color='green')
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
    axes[i].grid(True)  # Add grid lines

# Hide unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])  # Remove unused axes

plt.tight_layout()
plt.show()
```

```python
# Categorical Variables
categorical_columns = data.select_dtypes(include=['object']).columns

# Bar Charts for Categorical Variables
fig, axes = plt.subplots(len(categorical_columns) // 3 + 1, 3, figsize=(15, 15))
axes = axes.flatten()

# Iterate over categorical columns and plot
for i, col in enumerate(categorical_columns):
    sns.countplot(x=col, data=data, ax=axes[i], palette="Greens")
    axes[i].set_title(f'Bar Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].grid(True, axis='y', linestyle='--', alpha=0.7)  # Add grid lines only for the y-axis

# Hide unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])  # Remove unused axes

plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis
# Correlation Matrix
correlation_matrix = data[numerical_columns].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="Greens", fmt=".2f", vmin=-1, vmax=1)
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

```python
# Scatter Plot for Selected Pairs
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Thermo_Mid_Max', y='Thermo_Low_Max', hue='Burn_Treatment', palette="Greens")
plt.title("Scatter Plot: Thermo_Mid_Max vs Thermo_Low_Max")
plt.xlabel("Thermo_Mid_Max")
plt.ylabel("Thermo_Low_Max")
plt.legend(title="Burn Treatment")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D


# Box Plot for Numerical vs Categorical
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Burn_Treatment', y='Soil_pH', palette="Greens")
plt.title("Box Plot: Soil_pH by Burn Treatment")
plt.xlabel("Burn Treatment")
plt.ylabel("Soil pH")
plt.show()
```

```python
burn_counts = data['Burn_Treatment'].value_counts()
plt.figure(figsize=(8, 6))
burn_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('Greens', len(burn_counts)))
plt.title('Proportion of Burn Treatments', fontsize=14)
plt.ylabel('')
plt.show()
```

```python
# Box Plot for Numerical vs Categorical
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Burn_Treatment', y='Thermo_Mid_Max', palette="Greens")
plt.title("Box Plot: Thermo_Mid_Max by Burn Treatment")
plt.xlabel("Burn Treatment")
plt.ylabel("Thermo_Mid_Max")
plt.show()
```

```python
# Boxen Plot for Numerical vs Categorical
plt.figure(figsize=(10, 6))
sns.boxenplot(data=data, x='Burn_Treatment', y='Thermo_Low_Max', palette="Greens")
plt.title("Boxen Plot: Thermo_Low_Max by Burn Treatment")
plt.xlabel("Burn Treatment")
plt.ylabel("Thermo_Low_Max")
plt.show()
```

```python
# Pivot data for heatmap
heatmap_data = data.pivot_table(values='Thermo_Mid_Max', index='Burn_Treatment', columns='Terrain_Aspect', aggfunc='mean')

# Heatmap for aggregated data
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap="Greens", fmt=".2f")
plt.title("Heatmap: Mean Thermo_Mid_Max by Burn Treatment and Terrain Aspect")
plt.xlabel("Terrain Aspect")
plt.ylabel("Burn Treatment")
plt.show()
```

```python
# Categorical Scatter Plot
plt.figure(figsize=(10, 6))
sns.stripplot(data=data, x='Burn_Treatment', y='Soil_pH', jitter=True, palette="Greens")
plt.title("Categorical Scatter Plot: Soil pH by Burn Treatment")
plt.xlabel("Burn Treatment")
plt.ylabel("Soil pH")
plt.show()
```
