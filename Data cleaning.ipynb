```python
# Load your dataset
file_path = 'Updated_WoodBuffalo2019_Dataset.csv'
df = pd.read_csv(file_path)

# Function to identify outliers using IQR method
def detect_outliers_iqr(df):
    outliers = {}
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define the outlier bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Find outliers in the column
        outliers[column] = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    return outliers

# Function to handle outliers with mean and mode
def handle_outliers_iqr_mean_mode(df):
    df_handled = df.copy()  # Create a copy to avoid modifying the original DataFrame
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define the outlier bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Calculate mean and mode
        mean_value = df[column].mean()
        mode_value = df[column].mode()[0]  # mode() returns a Series, so we take the first value
        
        # Replace outliers with mean
        df_handled.loc[df_handled[column] < lower_bound, column] = mean_value  # Replace outliers below the lower bound
        df_handled.loc[df_handled[column] > upper_bound, column] = mean_value  # Replace outliers above the upper bound

        # Replace outliers with mode
        # df_handled.loc[df_handled[column] < lower_bound, column] = mode_value
        # df_handled.loc[df_handled[column] > upper_bound, column] = mode_value
        
    return df_handled

# Function to visualize comparisons
def visualize_comparison(df_original, df_handled):
    for column in df_original.select_dtypes(include=['float64', 'int64']).columns:
        plt.figure(figsize=(14, 6))

        # Boxplot before and after handling outliers
        plt.subplot(1, 2, 1)
        plt.boxplot([df_original[column], df_handled[column]], labels=['Original', 'Handled'])
        plt.title(f'Boxplot Comparison for {column}')
        plt.ylabel(column)

        # Histogram before and after handling outliers
        plt.subplot(1, 2, 2)
        plt.hist(df_original[column], bins=30, alpha=0.5, label='Original', edgecolor='black')
        plt.hist(df_handled[column], bins=30, alpha=0.5, label='Handled', edgecolor='black')
        plt.title(f'Histogram Comparison for {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.legend()

        plt.tight_layout()
        plt.show()

# 1. Identify outliers (not used directly for visualization)
outliers_iqr = detect_outliers_iqr(df)

# 2. Handle outliers using mean
df_handled_mean = handle_outliers_iqr_mean_mode(df)

# 3. Visualize comparison before and after handling outliers
visualize_comparison(df, df_handled_mean)
```
